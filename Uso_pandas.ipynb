```python
import miceforest as mf ## Cargamos la libreria miceforest
from sklearn.datasets import load_iris ## scextraigo el data set iris (lisbreria base de inteligencia artificial
import numpy as np # IMPORTAMOS NUMPY


```

```python
# cargando iris datasets
from sklearn import datasets
iris = datasets.load_iris()
iris
```

```python
# convertir iris (Skinlearn) en dataframe
#df_iris = pd.DataFrame(data=iris.data,
#                  columns=iris.feature_names)
## necsitamos la columna target

df_iris = pd.DataFrame(data=iris.data,
                  columns=iris.feature_names)
df_iris['target'] = iris.target
df_iris
```

```python
# REnombramos la columna target por species en su propio dataframe
df_iris.rename({"target": "species"}, inplace=True, axis=1)
df_iris
```

```python
# transformo el tipo de datos del campo species para categoria
df_iris['species'] = df_iris['species'].astype('category')
df_iris.info()
```

```python
## asigno valores nulos a la columna species aleatoriamente (Amputacion). se asiga el 0.25 (25%) de valores nulos 1025 profundidad del arbol
iris_amp = mf.ampute_data(df_iris, perc=0.25, random_state=1025)
iris_amp
```

```python
# aplicar la imputacion kernel

kds = mf.ImputationKernel(               ### nombre de dataframepara entrenar el modelo
    iris_amp, random_state=1991          ### Valores aleatorios para entrenar modelo
)
# corre el algoritmo MICE durante 2 interaciones
kds.mice(2)
# Retorna el DataSet completo
iris_completo = kds.complete_data()
iris_completo
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_hurto_motos = pd.read_csv('/content/drive/MyDrive/hurto_de_moto.csv')
df_hurto_personas = pd.read_csv('/content/drive/MyDrive/hurto_a_persona.csv')
```

```python
df_fecha = pd.Series(df_hurto_motos['fecha_hecho'])
df_fecha
```

```python
# damos un split
df_fecha2 = df_fecha.str.split('T') # convierte en una lista de 2 posiciones Fecha y hora

df_fecha2
```

```python
type(df_fecha2)
```

```python
#Convertimos de serie de Pansdas a DataFrame
df_fecha2 = pd.DataFrame(df_fecha2)
df_fecha2 = pd.DataFrame(df_fecha2.fecha_hecho.values.tolist(), index=df_fecha2.index)
df_fecha2
```

```python
type(df_fecha2)
```

```python
# Renombro las columnas
#df_fecha2 = df_fecha2.rename(columns={0:'Fecha'})
#df_fecha2 = df_fecha2.rename(columns={1:'Hora'})
df_fecha2

```

```python
# Transformarmos de DataFrame (String a formato Fecha)
df_fecha_datetime = pd.to_datetime(df_fecha2['Fecha'], format='%Y-%m-%d')
df_fecha_datetime
```

```python
df_fecha_datetime.min()
```

```python
df_fecha_datetime.max()
```

```python
#podemos hacer cosas comoe esta,  Identificamos el mes de cada registro
df_fecha_datetime_mes = df_fecha_datetime.dt.month
df_fecha_datetime_mes
```

```python
## identifico dias de la semana  Lunes=0 domingo=6
df_fecha_datetime_dia = df_fecha_datetime.dt.day_of_week
df_fecha_datetime_dia
```

```python
## hacer un histograma sobre los d√≠as utilizando la libreria seaborn
import seaborn as sns # Libreria para crear graficos estadisticos
import matplotlib.pyplot as plt # Libreria para crear graficos generales
```

```python
df_fecha_datetime_dia = pd.DataFrame(df_fecha_datetime_dia)
sns.histplot(data=df_fecha_datetime_dia, x='Fecha', stat='count')
#plt.xticks(rotation=90) #rota los nombres de las columnas cuando son muy largo, opcional
```

```python
df_fecha_datetime_mes = pd.DataFrame(df_fecha_datetime_mes)
df_fecha_datetime_mes
#pd.unique(df_fecha_datetime_mes)
sns.histplot(data=df_fecha_datetime_mes, x='Fecha', stat='count')


```

```python
df_fecha_datetime_year = df_fecha_datetime.dt.year
#df_fecha_datetime_year
df_fecha_datetime_year = pd.DataFrame(df_fecha_datetime_year)
#pd.unique(df_fecha_datetime_year)
sns.histplot(data=df_fecha_datetime_year, x='Fecha', stat='count')
plt.xticks(rotation=90)
```

```python
# Calcular Auto Correlacion de la serie Tiempo
# statsmodels Libreria aplicada a la estadisticas de autocorrelaci√≥n
from statsmodels.graphics.tsaplots import plot_acf
df_fecha_agr = df_fecha2.groupby('Fecha').size()
#df_fecha_agr
plt.figure(figsize=(12, 6))
plot_acf(df_fecha_agr) # you can adjust the numbers of lags as needed (lapso de dias a observar)
plt.xlabel('Lag')
plt.ylabel('AutoCorrelaci√≥n')
plt.title('AutoCorrelaci√≥n funci√≥n (ACF) Plot')
plt.show()
```
